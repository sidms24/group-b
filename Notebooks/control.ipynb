```python
#!/usr/bin/env python3
"""
Section 2.3.3: Control – Automated Per-(Method, Feature) Hyperparameter Tuning
================================================================================

For every combination of {MC, TD} × {feature representation}, this script:
  1. Sweeps batch sizes           → picks the best
  2. Sweeps alpha (LR) schedules  → picks the best (using best batch size)
  3. Sweeps epsilon schedules     → picks the best (using best batch size + alpha)
  4. Runs a final control run with the winning (batch_size, alpha, epsilon) triple
  
Outputs: figures in FIG_DIR, summary tables printed to stdout.
"""

import sys, os, warnings, itertools
sys.path.insert(0, '.')
warnings.filterwarnings('ignore')

import numpy as np
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
from tqdm import tqdm

from sections.episodes import generate_episodes, mse, huber, N_ACTIONS, CLIP
from sections.mc import MC
from sections.td import TD
from sections.sarsa import SARSA_lambda
from sections.features import (
    compute_features_batch_jax, N_FEATURES,
    compute_features_extended, N_FEATURES_EXTENDED,
    make_tile_coding_features,
)
from sections.schedules import (
    schedule_decay, schedule_cosine, schedule_cosine_warmup,
    epsilon_schedule_cosine, epsilon_schedule_exp,
)
from sections.evaluation import evaluate_theta, _winrate_episode
from sections.visualisations import (
    plot_learned_policy_grid, plot_policy_lines, plot_policy_bars,
    plot_feature_weights, plot_top_features,
)

SEED = 24
print(f"JAX devices: {jax.devices()}")



# 1.  GLOBALS

N           = 1_024_000          # total training episodes per run
EVAL_EVERY  = 10_240             # episodes between greedy evaluations
N_ACT       = 51                 # action discretisation
N_EVAL      = 2_000              # episodes for greedy eval during training
N_EVAL_WR   = 5_000              # episodes for win-rate evaluation
FIG_DIR     = "content/figures/control"
os.makedirs(FIG_DIR, exist_ok=True)

# Method-specific gradient clipping
MC_CLIP = 1000
TD_CLIP = 100



# 2.  FEATURE REPRESENTATIONS

tile_fn, tile_n_features = make_tile_coding_features(n_tilings=4, n_tiles=8)

def features_combined(v, t, own_b, opp_b, candidates):
    """Extended (65) + tile-coded → combined feature vector."""
    ext  = compute_features_extended(v, t, own_b, opp_b, candidates)
    tile = tile_fn(v, t, own_b, opp_b, candidates)
    return jnp.concatenate([ext, tile], axis=1)

COMBINED_N = N_FEATURES_EXTENDED + tile_n_features

FEATURE_CONFIGS = {
    f"Standard ({N_FEATURES})":         (compute_features_batch_jax, N_FEATURES),
    f"Extended ({N_FEATURES_EXTENDED})": (compute_features_extended,  N_FEATURES_EXTENDED),
    f"Tile ({tile_n_features})":         (tile_fn,                    tile_n_features),
    f"Combined ({COMBINED_N})":          (features_combined,          COMBINED_N),
}

print(f"Feature dims: standard={N_FEATURES}, extended={N_FEATURES_EXTENDED}, "
      f"tile={tile_n_features}, combined={COMBINED_N}")



# 3. HYPERPARAMETER SEARCH SPACES


# Batch sizes to try 
BATCH_SIZES = [1, 16, 64, 256, 1024]

# Alpha (LR) schedule candidates – INDEPENDENT per method 
MC_ALPHA_CONFIGS = {
    "cosine(2e-3→5e-5)":       lambda ep: schedule_cosine_warmup(ep, N, amax=2e-2,  amin=5e-6, wf=0.05),
    "cosine(5e-3→1e-5)":       lambda ep: schedule_cosine_warmup(ep, N, amax=5e-4,  amin=1e-5, wf=0.05),
    "cosine(1e-2→5e-5)":       lambda ep: schedule_cosine_warmup(ep, N, amax=1e-4,  amin=5e-6, wf=0.05),
    "cosine(5e-3→5e-6)":       lambda ep: schedule_cosine_warmup(ep, N, amax=5e-4,  amin=5e-8, wf=0.50),
    "decay(a0=1e-2, d=1e-2)":  lambda ep: schedule_decay(ep, a0=1e-3, d=1e-2),
}

TD_ALPHA_CONFIGS = {
    "cosine(5e-3→5e-6)":       lambda ep: schedule_cosine_warmup(ep, N, amax=5e-3,  amin=5e-6, wf=0.05),
    "cosine(1e-2→1e-5)":       lambda ep: schedule_cosine_warmup(ep, N, amax=1e-3,  amin=1e-5, wf=0.05),
    "cosine(2e-3→1e-6)":       lambda ep: schedule_cosine_warmup(ep, N, amax=2e-3,  amin=1e-6, wf=0.05),
    "cosine(5e-3→1e-4)":       lambda ep: schedule_cosine_warmup(ep, N, amax=5e-3,  amin=1e-8, wf=0.05),
    "decay(a0=5e-3, d=5e-3)":  lambda ep: schedule_decay(ep, a0=5e-3, d=5e-3),
}

# Epsilon schedule candidates (shared pool, swept per combo) 
EPS_CONFIGS = {
    "cosine 1.0→0.01":         lambda ep: epsilon_schedule_cosine(ep, N, emax=1.0, emin=0.01),
    "cosine 0.5→0.05":         lambda ep: epsilon_schedule_cosine(ep, N, emax=0.5, emin=0.05),
    "cosine 1.0→0.05":         lambda ep: epsilon_schedule_cosine(ep, N, emax=1.0, emin=0.05),
    "exp fast (d=0.999987)":   lambda ep: epsilon_schedule_exp(ep, es=1.0, ee=0.01, ed=0.999987),
    "exp slow (d=0.999991)":   lambda ep: epsilon_schedule_exp(ep, es=1.0, ee=0.01, ed=0.999991),
}

# Default alpha / epsilon used during sweeps of *other* axes
DEFAULT_ALPHA_FN = lambda ep: schedule_cosine_warmup(ep, N, amax=5e-2, amin=5e-5, wf=0.05)
DEFAULT_EPS_FN   = lambda ep: epsilon_schedule_cosine(ep, N, emax=1.0, emin=0.01)

# OVERIDES
MC_OVERRIDES = {"batch_size": 64}     # e.g. {"batch_size": 16}
TD_OVERRIDES = {"batch_size": 64}     # e.g. {"alpha_name": "cosine(5e-3→5e-6)"}



def converged_reward(res, frac=0.3):
    """Mean reward over the final `frac` of training."""
    n = max(1, int(len(res["rewards"]) * frac))
    return np.mean(res["rewards"][-n:])


def run_control(
    train_fn, train_kwargs, alpha_fn, eps_fn,
    N=N, eval_every=EVAL_EVERY, n_actions=N_ACT,
    n_features=N_FEATURES, feature_fn=compute_features_batch_jax,
    n_eval=N_EVAL, seed=SEED, label="",
):
    """Chunk-based on-policy control with warm-start theta and mini-batch SGD."""
    n_chunks = N // eval_every
    episodes = np.arange(N, dtype=np.float32)
    alphas_all = jnp.array([alpha_fn(int(e)) for e in episodes])
    eps_arr    = np.array([eps_fn(int(e)) for e in episodes], dtype=np.float32)
    alphas_chunked = alphas_all.reshape(n_chunks, eval_every)
    chunk_eps      = eps_arr[::eval_every][:n_chunks]

    gen_keys  = jax.random.split(jax.random.PRNGKey(seed), n_chunks)
    eval_keys = jax.random.split(jax.random.PRNGKey(seed + 30_000), n_chunks)

    theta = jnp.zeros(n_features)
    best_theta, best_reward = theta, -np.inf
    all_thetas, all_losses, all_rewards = [], [], []

    for chunk in tqdm(range(n_chunks), desc=label or "Control"):
        phis, rewards, _, _, _ = generate_episodes(
            eval_every, gen_keys[chunk],
            epsilon=float(chunk_eps[chunk]), theta=theta,
            prediction=False, n_actions=n_actions, feature_fn=feature_fn,
        )
        phis.block_until_ready()

        thetas_c, loss_c = train_fn(
            phis, rewards, alphas_chunked[chunk],
            eval_every=eval_every, theta_init=theta, **train_kwargs,
        )
        theta = thetas_c[-1]

        r = float(evaluate_theta(
            theta, eval_keys[chunk],
            n_eval=n_eval, n_actions=n_actions, feature_fn=feature_fn,
        ))
        if r > best_reward:
            best_reward = r
            best_theta  = theta

        all_thetas.append(np.array(theta))
        all_losses.append(float(loss_c[-1]))
        all_rewards.append(r)

    return dict(
        thetas=np.array(all_thetas), losses=np.array(all_losses),
        rewards=np.array(all_rewards), theta_final=all_thetas[-1],
        theta_best=np.array(best_theta), best_reward=best_reward,
    )


def eval_winrate(theta, key, n_eval, n_actions, feature_fn):
    """Per-game (total_reward, round_win_rate) arrays."""
    keys = jax.random.split(key, n_eval)
    return jax.vmap(lambda k: _winrate_episode(theta, k, n_actions, feature_fn))(keys)



# 6.  SCHEDULE VISUALISATION


def _plot_schedules(configs, title, ylabel, fname, yscale="log"):
    x = np.arange(N)
    fig, ax = plt.subplots(figsize=(10, 5))
    for name, fn in configs.items():
        ax.plot(x[::500], [fn(e) for e in x[::500]], label=name)
    ax.set_xlabel("Episode"); ax.set_ylabel(ylabel)
    if yscale:
        ax.set_yscale(yscale)
    ax.set_title(title); ax.legend(fontsize=8); ax.grid(True, alpha=0.3)
    plt.tight_layout(); plt.savefig(f"{FIG_DIR}/{fname}", bbox_inches="tight"); plt.show()

def _print_schedule_table(configs, label):
    print(f'\n{label:<30s} {"at ep 0":>10s} {"at 32k":>10s} {"at 64k":>10s} {"at 100k":>10s}')
    print("-" * 62)
    for name, fn in configs.items():
        print(f"{name:<30s} {fn(0):10.2e} {fn(32000):10.2e} {fn(64000):10.2e} {fn(99999):10.2e}")

print("\n── Schedule Visualisation ──")
_plot_schedules(MC_ALPHA_CONFIGS, "MC Alpha Schedules",  "α", "mc_alpha_schedules.png")
_plot_schedules(TD_ALPHA_CONFIGS, "TD Alpha Schedules",  "α", "td_alpha_schedules.png")
_plot_schedules(EPS_CONFIGS,      "Epsilon Schedules",   "ε", "epsilon_schedules.png", yscale=None)
_print_schedule_table(MC_ALPHA_CONFIGS, "MC Alpha Schedule")
_print_schedule_table(TD_ALPHA_CONFIGS, "TD Alpha Schedule")



# 7.  PER-(METHOD, FEATURE) HYPERPARAMETER SWEEP ENGINE

def _method_label(method_name, feat_name):
    return f"{method_name} | {feat_name}"


def sweep_batch_size(
    method_name, train_fn, clip, feat_name, feat_fn, n_feat,
    alpha_fn=DEFAULT_ALPHA_FN, eps_fn=DEFAULT_EPS_FN,
):
    """Sweep batch sizes for one (method, feature) combo.  Returns {bs: result}."""
    results = {}
    for bs in BATCH_SIZES:
        lbl = f"{method_name} {feat_name} bs={bs}"
        results[bs] = run_control(
            train_fn=train_fn,
            train_kwargs=dict(loss_fn=mse, reg=True, clip=clip, batch_size=bs),
            alpha_fn=alpha_fn, eps_fn=eps_fn,
            N=N, eval_every=EVAL_EVERY,
            n_features=n_feat, feature_fn=feat_fn,
            label=lbl,
        )
        print(f"  {lbl}: conv={converged_reward(results[bs]):.1f}")
    return results


def sweep_alpha(
    method_name, train_fn, clip, feat_name, feat_fn, n_feat,
    alpha_configs, best_bs, eps_fn=DEFAULT_EPS_FN,
):
    """Sweep alpha schedules for one (method, feature) combo.  Returns {name: result}."""
    results = {}
    for aname, afn in alpha_configs.items():
        lbl = f"{method_name} {feat_name} α={aname[:20]}"
        results[aname] = run_control(
            train_fn=train_fn,
            train_kwargs=dict(loss_fn=mse, reg=True, clip=clip, batch_size=best_bs),
            alpha_fn=afn, eps_fn=eps_fn,
            N=N, eval_every=EVAL_EVERY,
            n_features=n_feat, feature_fn=feat_fn,
            label=lbl,
        )
        print(f"  {lbl}: conv={converged_reward(results[aname]):.1f}")
    return results


def sweep_epsilon(
    method_name, train_fn, clip, feat_name, feat_fn, n_feat,
    best_bs, best_alpha_fn,
):
    """Sweep epsilon schedules for one (method, feature) combo.  Returns {name: result}."""
    results = {}
    for ename, efn in EPS_CONFIGS.items():
        lbl = f"{method_name} {feat_name} ε={ename}"
        results[ename] = run_control(
            train_fn=train_fn,
            train_kwargs=dict(loss_fn=mse, reg=True, clip=clip, batch_size=best_bs),
            alpha_fn=best_alpha_fn, eps_fn=efn,
            N=N, eval_every=EVAL_EVERY,
            n_features=n_feat, feature_fn=feat_fn,
            label=lbl,
        )
        print(f"  {lbl}: conv={converged_reward(results[ename]):.1f}")
    return results


def find_best_hyperparams(method_name, train_fn, clip, alpha_configs, overrides,
                          feat_name, feat_fn, n_feat):
    """
    Full 3-stage sequential sweep for one (method, feature) combo.
    Returns (best_bs, best_alpha_name, best_alpha_fn, best_eps_name, best_eps_fn,
             sweep_details_dict).

    Any key present in `overrides` skips the sweep for that axis and uses the
    forced value instead.
    """
    details = {}

    # Stage 1: Batch Size 
    if "batch_size" in overrides:
        best_bs = overrides["batch_size"]
        print(f"  [OVERRIDE] batch_size = {best_bs}")
        details["bs_sweep"] = None
    else:
        print(f"\n  ── Sweeping batch sizes for {method_name} | {feat_name} ──")
        bs_results = sweep_batch_size(
            method_name, train_fn, clip, feat_name, feat_fn, n_feat)
        best_bs = max(bs_results, key=lambda k: converged_reward(bs_results[k]))
        details["bs_sweep"] = bs_results
    print(f"  → Best batch_size = {best_bs}")

    # Stage 2: Alpha Schedule
    if "alpha_name" in overrides:
        best_alpha_name = overrides["alpha_name"]
        best_alpha_fn   = alpha_configs[best_alpha_name]
        print(f"  [OVERRIDE] alpha = {best_alpha_name}")
        details["alpha_sweep"] = None
    else:
        print(f"\n  ── Sweeping alpha for {method_name} | {feat_name} (bs={best_bs}) ──")
        alpha_results = sweep_alpha(
            method_name, train_fn, clip, feat_name, feat_fn, n_feat,
            alpha_configs, best_bs)
        best_alpha_name = max(alpha_results, key=lambda k: converged_reward(alpha_results[k]))
        best_alpha_fn   = alpha_configs[best_alpha_name]
        details["alpha_sweep"] = alpha_results
    print(f"  → Best alpha = {best_alpha_name}")

    #  Stage 3: Epsilon Schedule
    if "eps_name" in overrides:
        best_eps_name = overrides["eps_name"]
        best_eps_fn   = EPS_CONFIGS[best_eps_name]
        print(f"  [OVERRIDE] epsilon = {best_eps_name}")
        details["eps_sweep"] = None
    else:
        print(f"\n  ── Sweeping epsilon for {method_name} | {feat_name} "
              f"(bs={best_bs}, α={best_alpha_name}) ──")
        eps_results = sweep_epsilon(
            method_name, train_fn, clip, feat_name, feat_fn, n_feat,
            best_bs, best_alpha_fn)
        best_eps_name = max(eps_results, key=lambda k: converged_reward(eps_results[k]))
        best_eps_fn   = EPS_CONFIGS[best_eps_name]
        details["eps_sweep"] = eps_results
    print(f"  → Best epsilon = {best_eps_name}")

    return best_bs, best_alpha_name, best_alpha_fn, best_eps_name, best_eps_fn, details



# 8.  MAIN SWEEP iterate over {MC,TD} × feature_configs


METHOD_SPECS = {
    "MC": dict(train_fn=MC, clip=MC_CLIP, alpha_configs=MC_ALPHA_CONFIGS, overrides=MC_OVERRIDES),
    "TD": dict(train_fn=TD, clip=TD_CLIP, alpha_configs=TD_ALPHA_CONFIGS, overrides=TD_OVERRIDES),
}

# Stores: combo_key → {best_bs, best_alpha_name, best_alpha_fn,
#                       best_eps_name, best_eps_fn, sweep_details}
best_hparams = {}

for method_name, mspec in METHOD_SPECS.items():
    for feat_name, (feat_fn, n_feat) in FEATURE_CONFIGS.items():
        combo_key = _method_label(method_name, feat_name)
        print(f"\n{'='*80}")
        print(f"  TUNING: {combo_key}")
        print(f"{'='*80}")

        bs, aname, afn, ename, efn, details = find_best_hyperparams(
            method_name=method_name,
            train_fn=mspec["train_fn"],
            clip=mspec["clip"],
            alpha_configs=mspec["alpha_configs"],
            overrides=mspec["overrides"],
            feat_name=feat_name,
            feat_fn=feat_fn,
            n_feat=n_feat,
        )
        best_hparams[combo_key] = dict(
            method=method_name, feat_name=feat_name,
            feat_fn=feat_fn, n_feat=n_feat,
            best_bs=bs,
            best_alpha_name=aname, best_alpha_fn=afn,
            best_eps_name=ename,   best_eps_fn=efn,
            sweep_details=details,
        )

# Print discovered hyperparameter summary 
print(f"\n\n{'='*90}")
print("  HYPERPARAMETER SUMMARY  (per method × feature)")
print(f"{'='*90}")
print(f"{'Combo':<40s} {'BS':>5s}  {'Alpha Schedule':<30s}  {'Epsilon Schedule':<25s}")
print("-" * 105)
for combo_key, hp in best_hparams.items():
    print(f"{combo_key:<40s} {hp['best_bs']:>5d}  "
          f"{hp['best_alpha_name']:<30s}  {hp['best_eps_name']:<25s}")


# 9.  FINAL RUNS  –  train each combo with its best hyperparameters

final_results = {}   # combo_key → run_control result

print(f"\n\n{'='*80}")
print("  FINAL TRAINING RUNS  (best hyperparameters per combo)")
print(f"{'='*80}\n")

for combo_key, hp in best_hparams.items():
    mspec = METHOD_SPECS[hp["method"]]
    lbl   = f"FINAL {combo_key}"
    print(f"── {lbl}  (bs={hp['best_bs']}, α={hp['best_alpha_name']}, "
          f"ε={hp['best_eps_name']}) ──")

    final_results[combo_key] = run_control(
        train_fn=mspec["train_fn"],
        train_kwargs=dict(loss_fn=mse, reg=True, clip=mspec["clip"],
                          batch_size=hp["best_bs"]),
        alpha_fn=hp["best_alpha_fn"],
        eps_fn=hp["best_eps_fn"],
        N=N, eval_every=EVAL_EVERY,
        n_features=hp["n_feat"], feature_fn=hp["feat_fn"],
        label=lbl,
    )
    res = final_results[combo_key]
    print(f"  conv={converged_reward(res):.1f}, best={res['best_reward']:.1f}\n")



# 10. COMPARISON PLOTS  –  reward curves, bar chart, summary table


x_ep = np.arange(N // EVAL_EVERY) * EVAL_EVERY
palette = ['#2196F3', '#FF5722', '#4CAF50', '#9C27B0', '#FF9800',
           '#00BCD4', '#E91E63', '#8BC34A']

# 10a. Reward curves: one figure per method 
for method_name in METHOD_SPECS:
    fig, ax = plt.subplots(figsize=(12, 5))
    combos = [(k, v) for k, v in final_results.items() if k.startswith(method_name)]
    for i, (combo_key, res) in enumerate(combos):
        ax.plot(x_ep, res["rewards"], label=combo_key, color=palette[i % len(palette)], alpha=0.8)
    ax.set_title(f"{method_name}: Final Runs – Greedy Reward by Feature Repr.")
    ax.set_xlabel("Episodes"); ax.set_ylabel("Greedy Reward")
    ax.legend(fontsize=7); ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(f"{FIG_DIR}/{method_name.lower()}_final_feature_comparison.png", bbox_inches="tight")
    plt.show()

#  10b. Combined bar chart
all_names = list(final_results.keys())
all_convs = [converged_reward(final_results[k]) for k in all_names]
bar_colors = ["#2196F3" if "MC" in n else "#FF5722" for n in all_names]

fig, ax = plt.subplots(figsize=(max(14, len(all_names) * 1.5), 6))
x_bar = np.arange(len(all_names))
bars  = ax.bar(x_bar, all_convs, color=bar_colors, alpha=0.85, edgecolor="white")
ax.set_xticks(x_bar)
ax.set_xticklabels(all_names, rotation=45, ha="right", fontsize=7)
ax.set_ylabel("Converged Reward (last 30%)")
ax.set_title("Final Runs: Converged Reward per (Method × Feature)")
ax.grid(True, axis="y", alpha=0.3)
for bar, val in zip(bars, all_convs):
    ax.text(bar.get_x() + bar.get_width() / 2, val + 0.5, f"{val:.1f}",
            ha="center", va="bottom", fontsize=7)
plt.tight_layout()
plt.savefig(f"{FIG_DIR}/final_bar_chart.png", bbox_inches="tight")
plt.show()

# 10c. Summary table
print(f"\n{'='*100}")
print("  FINAL RESULTS SUMMARY")
print(f"{'='*100}")
print(f"{'Combo':<40s} {'BS':>4s} {'Alpha':<25s} {'Epsilon':<22s} "
      f"{'Conv':>8s} {'Best':>8s}")
print("-" * 112)
for combo_key in final_results:
    hp  = best_hparams[combo_key]
    res = final_results[combo_key]
    print(f"{combo_key:<40s} {hp['best_bs']:>4d} {hp['best_alpha_name']:<25s} "
          f"{hp['best_eps_name']:<22s} {converged_reward(res):>8.2f} "
          f"{res['best_reward']:>8.2f}")

overall_best_key = max(final_results, key=lambda k: converged_reward(final_results[k]))
print(f"\n  OVERALL BEST: {overall_best_key}  "
      f"(conv={converged_reward(final_results[overall_best_key]):.2f}, "
      f"peak={final_results[overall_best_key]['best_reward']:.2f})")



# 11. WIN-RATE EVALUATION  –  all final models vs default Player 2

print(f"\n\nEvaluating {len(final_results)} final agents over {N_EVAL_WR} games each...\n")
wr_master_key = jax.random.PRNGKey(SEED + 200)
all_wr = {}

for combo_key, res in tqdm(final_results.items(), desc="Win-rate eval"):
    hp = best_hparams[combo_key]
    wr_master_key, subkey = jax.random.split(wr_master_key)
    theta_jnp = jnp.array(res["theta_best"])
    rewards, round_wrs = jax.jit(
        eval_winrate, static_argnames=("n_eval", "n_actions", "feature_fn")
    )(theta_jnp, subkey, n_eval=N_EVAL_WR, n_actions=N_ACT, feature_fn=hp["feat_fn"])
    rw_np, rr_np = np.array(rewards), np.array(round_wrs)
    all_wr[combo_key] = {
        "game_wr":    (rw_np > 75.0).mean(),
        "round_wr":   rr_np.mean(),
        "mean_r":     rw_np.mean(),
        "std_r":      rw_np.std(),
        "median_r":   np.median(rw_np),
        "rewards_np": rw_np,
        "round_wrs_np": rr_np,
    }

# Print win-rate table 
sorted_wr = sorted(all_wr.items(), key=lambda x: -x[1]["game_wr"])
print(f'\n{"#":<4s} {"Agent":<40s} {"Game WR":>8s} {"Round WR":>9s} '
      f'{"Mean R":>8s} {"Std R":>7s} {"Median":>8s}')
print("=" * 86)
for i, (name, r) in enumerate(sorted_wr, 1):
    print(f'{i:<4d} {name:<40s} {r["game_wr"]:8.1%} {r["round_wr"]:9.1%} '
          f'{r["mean_r"]:8.1f} {r["std_r"]:7.1f} {r["median_r"]:8.1f}')

# Win-rate bar chart (top 8) 
top_n     = sorted_wr[:min(8, len(sorted_wr))]
names_top = [n for n, _ in top_n]
gwrs_top  = [r["game_wr"]  for _, r in top_n]
rwrs_top  = [r["round_wr"] for _, r in top_n]
colors_top = ["#2196F3" if "MC" in n else "#FF5722" for n in names_top]

fig, ax = plt.subplots(figsize=(14, 6))
x_pos = np.arange(len(names_top))
w = 0.35
ax.bar(x_pos - w/2, gwrs_top, w, color=colors_top, alpha=0.85, label="Game WR")
ax.bar(x_pos + w/2, rwrs_top, w,
       color=[c.replace("F3", "99").replace("22", "88") for c in colors_top],
       alpha=0.6, label="Round WR")
ax.axhline(0.5, color="gray", ls="--", alpha=0.5)
ax.set_xticks(x_pos)
ax.set_xticklabels(names_top, rotation=40, ha="right", fontsize=7)
ax.set_ylabel("Win Rate")
ax.set_title("Top Agents by Game Win Rate vs Default Player 2")
ax.legend(); ax.grid(True, axis="y", alpha=0.3)
for i, (gw, rw) in enumerate(zip(gwrs_top, rwrs_top)):
    ax.text(i - w/2, gw + 0.008, f"{gw:.0%}", ha="center", va="bottom", fontsize=7)
plt.tight_layout()
plt.savefig(f"{FIG_DIR}/final_winrate_top.png", bbox_inches="tight")
plt.show()

# Reward distribution histograms 
fig, ax = plt.subplots(figsize=(12, 5))
for i, (name, r) in enumerate(sorted_wr[:4]):
    c = palette[i % len(palette)]
    ax.hist(r["rewards_np"], bins=40, alpha=0.4, color=c, label=name, edgecolor="white")
    ax.axvline(r["mean_r"], color=c, ls="--", lw=1.5)
ax.axvline(75, color="gray", ls=":", alpha=0.5, label="Break-even (75)")
ax.set_xlabel("Total Game Reward"); ax.set_ylabel("Count")
ax.set_title("Reward Distribution – Top 4 Agents")
ax.legend(fontsize=7); ax.grid(True, alpha=0.2)
plt.tight_layout()
plt.savefig(f"{FIG_DIR}/final_reward_distributions.png", bbox_inches="tight")
plt.show()



# 12. LEARNED STRATEGY VISUALISATION  –  overall best agent


overall_best_res = final_results[overall_best_key]
overall_hp       = best_hparams[overall_best_key]
best_theta       = jnp.array(overall_best_res["theta_best"])

print(f"\nVisualising strategy for: {overall_best_key}")
print(f"  conv={converged_reward(overall_best_res):.2f}, "
      f"peak={overall_best_res['best_reward']:.2f}")

plot_learned_policy_grid(best_theta, title_prefix=overall_best_key,
                         n_actions=N_ACT, feature_fn=overall_hp["feat_fn"])
plot_policy_lines(best_theta, title=f"{overall_best_key}: Bid Fraction vs Prize Value",
                  n_actions=N_ACT, feature_fn=overall_hp["feat_fn"])
plot_policy_bars(best_theta, title=overall_best_key,
                 n_actions=N_ACT, feature_fn=overall_hp["feat_fn"])

# Feature weight analysis (only meaningful for non-tile features)
if overall_hp["n_feat"] <= N_FEATURES_EXTENDED + 50:
    plot_top_features(overall_best_res["theta_best"], title=overall_best_key)
    plot_feature_weights(best_theta, title=overall_best_key)
else:
    print(f"  (Skipping per-feature-weight plots for {overall_hp['n_feat']}-dim representation)")



# 13. GRAND SUMMARY


print(f"\n\n{'='*100}")
print("  GRAND SUMMARY")
print(f"{'='*100}")
print(f"N={N:,}, n_actions={N_ACT}, eval_every={EVAL_EVERY}")
print(f"MC clip={MC_CLIP}, TD clip={TD_CLIP}")
print(f"MC overrides: {MC_OVERRIDES if MC_OVERRIDES else 'None (all swept)'}")
print(f"TD overrides: {TD_OVERRIDES if TD_OVERRIDES else 'None (all swept)'}")
print()

print(f"{'Combo':<40s} {'BS':>4s} {'Alpha':<25s} {'Eps':<22s} "
      f"{'Conv':>8s} {'Best':>8s} {'GameWR':>8s}")
print("-" * 120)
for combo_key in final_results:
    hp  = best_hparams[combo_key]
    res = final_results[combo_key]
    wr  = all_wr[combo_key]
    print(f"{combo_key:<40s} {hp['best_bs']:>4d} {hp['best_alpha_name']:<25s} "
          f"{hp['best_eps_name']:<22s} {converged_reward(res):>8.2f} "
          f"{res['best_reward']:>8.2f} {wr['game_wr']:>8.1%}")

wr_best = sorted_wr[0]
print(f"\n  OVERALL BEST (by win rate): {wr_best[0]}")
print(f"    Game WR = {wr_best[1]['game_wr']:.1%},  Mean Reward = {wr_best[1]['mean_r']:.1f}")
print(f"\n  OVERALL BEST (by conv reward): {overall_best_key}")
print(f"    Conv = {converged_reward(final_results[overall_best_key]):.2f},  "
      f"Peak = {final_results[overall_best_key]['best_reward']:.2f}")
print(f"\nAll figures saved to {FIG_DIR}/")
```
